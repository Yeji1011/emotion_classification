```python
import torch
import torchvision
import torch.nn # as nn
import numpy as np
import os
from PIL import Image
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

if torch.cuda.is_available():
    print("------- GPU Working -------")
    print("[Current GPU]: " + str(torch.cuda.get_device_name(0)))
else:
    print("------- CPU Working -------")
print('[PyTorch version]: ', torch.__version__)
```

```python
def load_checkpoint(subject):
    # batch_size=1
    num_classes=7
    
    model=torchvision.models.resnet152(pretrained=True)

    model_dir_path = 'C:\\Users\\yj\\Desktop/Task/Trasnfer_Learning/Resnet152/Best_model_state'
    model_save_dir = os.path.join(model_dir_path, '{}_model.pth'.format(subject))
    
    checkpoint = torch.load(model_save_dir)

    num_ftrs = model.fc.in_features #체크포인트 로드하기전에 동일하게(마지막단 노드수 7) 변경
    model.fc = torch.nn.Linear(num_ftrs, num_classes) #(fc): Linear(in_features=2048, out_features=7, bias=True)
    model.to(device)
    model.load_state_dict(checkpoint)

    model.eval()
    
    return model
    
    print('Model loaded!')
```

```python
# 디렉토리안 파일들(풀파일주소) 가져오기
import os

dirname='C:\\Users\\yj\\Desktop/Dataset/YFace_DB_edit/YFace_DB_edit'

def search(dirname):    
    filenames = os.listdir(dirname) #['.ipynb_checkpoints', 'F10_AN_O.png', #listdir 디렉토리안 모든 파일명 가져옴, 앞에 주소빼고
    
    filepath_set =[] #풀파일명(파일주소) 담기
    for filename in filenames:
        full_filename = os.path.join(dirname, filename)   #C:\Users\yj\Desktop/Dataset/YFace_DB_edit/YFace_DB_edit\F10_AN_O.png #지금주소랑 디렉토리안 파일명 합쳐서 파일주소만들기
        ext = os.path.splitext(full_filename)[-1]#.png  #splittext는 확장자로 스플릿 -> 맨끝(.png)를 ext로 저장
        if ext == '.png': #맨끝(확장자가) png이면 
            filepath_set.append(full_filename)  #['C:\\Users\\yj\\Desktop/Dataset/YFace_DB_edit/YFace_DB_edit\\F10_AN_O.png']
    return filepath_set
   
filepath_set = search(dirname)  #len(filename_set) 519
```

```python
from collections import defaultdict


def classfy_files_by_subject(filepath_set):
    
    files_by_subject = defaultdict(list) #디폴트값이 list인 딕셔너리

    for file_name in filepath_set:

        image_name = file_name.split("\\")   #F10_AN_O.png
        num_subject= image_name[-1][:-9]  #F10
        
        # emotion_name = file_name[-8:-6]   #AN

        files_by_subject[num_subject].append(file_name)      
    return files_by_subject

files_by_subject =classfy_files_by_subject(filepath_set)
```

```python
from collections import defaultdict

def feature_extraction(num_subject, emotion_name, file_path):
    
    # ft_by_subject = defaultdict(list) #디폴트값이 list인 딕셔너리
    
    model = load_checkpoint(num_subject)
    
    image = Image.open(file_path)
    image = image.convert('RGB')
    # plt.imshow(image)
    valid_transform = torchvision.transforms.Compose([
                                    torchvision.transforms.Resize((224, 224)),
                                    torchvision.transforms.ToTensor(),
                                    torchvision.transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
                                    ])
    X = valid_transform(image)
    X = X.unsqueeze(0)
    X = X.to(device)
    # print('image size: ', X.size())
    # model.eval()
    
    activation = {}
    def getActivation(name):
        # the hook signature
        def hook(model, input, output):
            activation[name] = output.detach()
        return hook

    h1 = model.avgpool.register_forward_hook(getActivation('avgpool'))
    out = model(X)
    h1.remove()
    
    after_avgpool = activation['avgpool']
    after_flatten = torch.flatten(after_avgpool, 1)
    out = after_flatten.cpu().numpy() 
    # print('{}_feature size:'.format(emotion_name), out.size())
    # out = out.tolist()
    # print('{} feature:'.format(emotion_name), out)
    # print(num_subject, emotion_name, out)
    return out
```
